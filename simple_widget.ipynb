# Exemplo rápido de Widgets

Este notebook demonstra vários controles do `ipywidgets` para você testar: RadioButtons, Checkbox, ToggleButtons, Dropdown, SelectMultiple, Sliders, Buttons e como capturar/mostrar valores.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

ingredients_df = pd.read_csv('./src/ingredients_dataset.csv')
n_ingredientes = len(ingredients_df)

# aumentar largura do rótulo (description) e do controle
_orig_intslider = widgets.IntSlider
def _styled_intslider(*args, **kwargs):
    style = dict(kwargs.get('style', {}))
    style.setdefault('description_width', '220px')
    kwargs['style'] = style
    kwargs.setdefault('layout', widgets.Layout(width='600px'))
    return _orig_intslider(*args, **kwargs)
widgets.IntSlider = _styled_intslider

slides = []
for i in range(n_ingredientes):
    ingrediente = ingredients_df.iloc[i]['name']
    min_value = ingredients_df.iloc[i]['min_value']
    max_value = ingredients_df.iloc[i]['max_value']
    min_default = ingredients_df.iloc[i]['min_default']
    slides.append(widgets.IntSlider( value=min_default, min=min_value, max=max_value, step=1, description=ingrediente))

out = widgets.Output()

def on_click(b):
    with out:
        clear_output()
        for i, slide in enumerate(slides):
            ingrediente = ingredients_df.iloc[i]['name']
            print(f'{ingrediente}: {slide.value}')

button = widgets.Button(description='Mostrar', button_style='primary')
button.on_click(on_click)
display(widgets.VBox([widgets.VBox(slides), button, out]))


```
